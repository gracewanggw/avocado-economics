```python
dfEcon = pd.read_csv('./macroeconomics.csv')
dfEcon
```

```python
dfEcon['date'] = pd.to_datetime(dfEcon['date'])
dfEcon['year'] = [date.year for date in dfEcon['date']]
dfEcon = dfEcon.where(dfEcon['year'] >= 2015).dropna()
dfEcon = dfEcon.where(dfEcon['year'] <= 2018).dropna()
dfEcon
```

```python
econ_cleaned = pd.DataFrame()
econ_cleaned['date'] = dfEcon['date']
econ_cleaned['GDP'] = dfEcon['Gross Domestic Product']
econ_cleaned['GNP'] = dfEcon['Gross National Product']
econ_cleaned['Unemployment Rate'] = dfEcon['Unemployment Rate']
econ_cleaned['Labor Force Participation Rate'] = dfEcon['Labor Force Participation Rate']
econ_cleaned['Personal Consumption Expenditures'] = dfEcon['Personal Consumption Expenditures']
econ_cleaned['Trade Balance'] = dfEcon['Trade Balance: Goods and Services, Balance of Payments Basis']
econ_cleaned['Real Export'] = dfEcon['Real Exports of Goods and Services']
econ_cleaned['Personal Income'] = dfEcon['Disposable Personal Income']
econ_cleaned['Personal Saving'] = dfEcon['Personal Saving']
econ_cleaned
```

```python
# Mapping each row to calendar quarter
econ_cleaned['date'] = pd.to_datetime(econ_cleaned['date'])
quarter = []
year = []
for date in econ_cleaned['date']:
    year.append(date.year)
    if date.month == 4:
        quarter.append('Q1')
    elif date.month == 7:
        quarter.append('Q2')
    elif date.month == 10:
        quarter.append('Q3')
    else:
        quarter.append('Q4')
econ_cleaned['quarter'] = quarter
econ_cleaned['year'] = year
econ_cleaned = econ_cleaned.drop('date', axis=1)
econ_cleaned
```

```python
# Sort the DataFrame by the 'YearQuarter' column
econ_cleaned['YearQuarter'] = econ_cleaned['year'].astype(str) + econ_cleaned['quarter']
econ_cleaned = econ_cleaned.sort_values(by='YearQuarter')
econ_cleaned.drop(['YearQuarter'], axis=1, inplace=True)
```

```python
quarter_to_months = {
    'Q1': ['01', '02', '03'],
    'Q2': ['04', '05', '06'],
    'Q3': ['07', '08', '09'],
    'Q4': ['10', '11', '12']
}

dates = []
for i, row in econ_cleaned.iterrows():
    for mnth in quarter_to_months[row['quarter']]:
        dates.append( str(row['year']) + '-' + mnth + '-01' )
```

```python
# There will be a row for each month in the quarter so we have to duplicate
econ_cleaned = pd.concat([econ_cleaned] * 3, ignore_index=True)
```

```python
# Sort the DataFrame by the 'YearQuarter' column
econ_cleaned['YearQuarter'] = econ_cleaned['year'].astype(str) + econ_cleaned['quarter']
econ_cleaned = econ_cleaned.sort_values(by='YearQuarter')
econ_cleaned.drop(['YearQuarter'], axis=1, inplace=True)
```

```python
econ_cleaned['date'] = dates
econ_cleaned
```

```python
econ_cleaned.to_csv('./macroCleaned.csv')
```
